Implementing Logistic Regression

```python
import pandas
```

```python
df=pandas.read_csv('/content/sample_data/car_data.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df
```

```python
dict={'Male':1,'Female':0}
df['Gender']=df['Gender'].map(dict)
```

```python
df.columns.values
```

```python
df.describe()
```

```python
x=list(df.columns.values[:-1])
```

```python
x
```

```python
y=df.columns.values[-1]
```

```python
y
```

```python
train_data_x,test_data_x,train_data_y,test_data_y=train_test_split(df[x],df[y],train_size=0.8,test_size=0.2,random_state=0)
```

```python
train_data_x.shape
```

```python
train_data_y.shape
```

```python
test_data_x.shape
```

```python
test_data_y.shape
```

```python
classi=LogisticRegression().fit(train_data_x,train_data_y)
```

```python
specific=['Age','AnnualSalary']
```

```python
classi1=LogisticRegression().fit(train_data_x[specific],train_data_y)
```

Accuracy of train data with all features

```python
classi.score(train_data_x,train_data_y)
```

Acuuracy of train data with few features

```python
classi1.score(train_data_x[specific],train_data_y)
```

```python
model1=classi.predict(test_data_x)
```

```python
from sklearn import metrics
```

Accuracy of test data with all features

```python
metrics.accuracy_score(test_data_y,model1)
```

```python
model2=classi1.predict(test_data_x[specific])
```

Accuracy of test data with few features

```python
metrics.accuracy_score(test_data_y,model2)
```
